# Cumilative Distribution Function (CDF)

```python
# Calculation CDF for Sleep Duration
def calculate_cdf(sleep_dur_list):
  sorted_data = sorted(sleep_dur_list)
  return sorted_data, np.arange(0,1,1/len(sorted_data))

x, y = calculate_cdf(sleep_dur_list)
plt.plot(x,y, label="CDF", color='red')
plt.xlabel('x')
plt.ylabel('F^(x)')
plt.legend()
```

```python
# Inverse CDF for Sleep Duration
def quartiles(sleep_dur_list):
  Q0 = sleep_dur_list[0] # 0%
  Q1 = sleep_dur_list[int(len(sleep_dur_list)*0.25)] # 25%
  Q2 = sleep_dur_list[int(len(sleep_dur_list)*0.5)] # 50%
  Q3 = sleep_dur_list[int(len(sleep_dur_list)*0.75)] # 75%
  Q4 = sleep_dur_list[int(len(sleep_dur_list)*1.0)-1] # 100%

  return Q0,Q1,Q2, Q3, Q4
Q = quartiles(x)
Q
```

```python
# Inverse CDF Sleep Duration
plt.plot(y,x, label = "Inverse CDF")
plt.scatter([0,0.25,0.5,0.75,1], Q, color = 'r') # 0.5

#plot.plot()
plt.xlabel('Probability')
plt.ylabel('F^-1(p)')
plt.legend()
```

```python
# IQR
IQR = Q[3]-Q[1]
IQR
```

```python
# Outlier detection
less_min = df[df['Sleep Duration']<Q[1]-1.5*IQR] # less than min
more_max = df[df['Sleep Duration']>Q[1]+1.5*IQR] # less than min
print(less_min.size)
print(more_max.size)
```

```python
#PMF for Sleep Duration

def calculate_pmf(sleep_dur_list):
  unique_values, counts = np.unique(sleep_dur_list, return_counts=True)
  pmf = counts/len(sleep_dur_list)
  return unique_values, pmf

# Example usage (assuming 'df' is your DataFrame)
unique_x, pmf_y = calculate_pmf(df['Sleep Duration'])

#Plotting the PMF
plt.stem(unique_x, pmf_y, label='PMF')
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.legend()
```

```python
# Calculating Variance for Sleep Duration
def variance(array):
  # mean
  mean = np.mean(array)
  #farqlar kvadrati
  sq_diff = (array - mean)**2

  #variance
  return np.sum(sq_diff)/len(array)

variance = variance(df['Sleep Duration'].values)
variance
```

```python
# Bivariate analysis with Sleep Duration and Quality of Sleep

# for Sleep Duration
sleep_dur_list = df['Sleep Duration'].values.tolist()
sleep_dur_array = df['Sleep Duration'].values
sleep_dur_frame = df['Sleep Duration']

# for Quality of Sleep
quality_sleep_list = df['Quality of Sleep'].values.tolist()
quality_sleep_array = df['Quality of Sleep'].values
quality_sleep_frame = df['Quality of Sleep']
```

```python
# Calculating coveriance for Sleep Duration and Quality of Sleep
def calculate_covariance(sleep_dur_list,quality_sleep_list):
  if len(sleep_dur_list) != len(quality_sleep_list):
    raise ValueError("List x1 va x2 should be equal each other")

  n = len(sleep_dur_list)
  mean_sleep_dur = sum(sleep_dur_list)/n
  mean_quality_sleep = sum(quality_sleep_list)/n

  covariance = sum((sleep_dur_list[i]-mean_sleep_dur)*(quality_sleep_list[i]-mean_quality_sleep) for i in range(n))/n
  return covariance

calculate_covariance(sleep_dur_list, quality_sleep_list)
```

```python
# Normal distribution for Sleep Duration

def calculate_pdf_vectorized(data):
  mu = np.mean(data)
  sigma = np.std(data, ddof=1)
  coeff = 1/(sigma*np.sqrt(2*np.pi))

  # Compute the exponent part for all data points in one go
  exponent = np.exp(-((data-mu)**2)/(2*sigma**2))

  return coeff*exponent

# Calculate for PDF for all data in sepal length array

pdf_values = calculate_pdf_vectorized(sleep_dur_array)
print("PDF values for the sepal length data:")
print(pdf_values)
plt.scatter(sleep_dur_array, pdf_values)
```

```python
# Calculating correlation for Sleep Duration and Quality of Sleep

def calculate_correlation(X1, X2):
  if len(X1)!= len(X2):
    raise ValueError("List x1 va x2 bir biriga teng bo'lishi kerak")

  n = len(X1)
  mean_X1 = sum(X1)/n
  mean_X2 = sum(X2)/n

  #calculate covariance

  cov = sum((X1[i]-mean_X1)*(X2[i]-mean_X2) for i in range(n))/n

  #calculate  the SD of X1 and X2
  std_X1 = (sum((xi-mean_X1)**2 for xi in X1)/n)**0.5
  std_X2 = (sum((xi-mean_X2)**2 for xi in X2)/n)**0.5

  #Calculate correlation coofeitsient
  if std_X1 == 0 or std_X2 == 0:
    raise ValueError("SD cannot be zero for correlaation calculation")

  correlation = cov/(std_X1*std_X2)

  return correlation

calculate_correlation(sleep_dur_list, quality_sleep_list)
```
